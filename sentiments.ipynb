```python
# After downloading Wikipedia and langdetect, we start downloading the various libraries.

import bs4 as bs
from bs4 import BeautifulSoup
import pandas as pd
import urllib.request as urllib2
from urllib.request import urlopen
import re
import nltk
import wikipedia
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords

nltk.download('stopwords')
nltk.download('punkt')
```

```python
# In the first exercise, we will create a wordcloud of the "Paracetamol" entry from the German Wikipedia page.

import urllib.request

def cargar_texto(url):
    scraped_data = urllib.request.urlopen(url)
    article = scraped_data.read()
    parsed_article = bs.BeautifulSoup(article,'lxml')
    paragraphs = parsed_article.find_all('p')
    article_text = ""
    for p in paragraphs:
        article_text += p.text
    return article_text

def hacer_resumen(article_text):
    summary = ""
    return summary

article_text = cargar_texto('https://de.wikipedia.org/wiki/Paracetamol')
resumen = hacer_resumen(article_text)
print(resumen)
```

```python
cloud_article_text = WordCloud().generate(article_text)
```

```python
# In the second exercise, we will incorporate various elements of difficulty. Therefore, we will create four wordclouds (Catalan, English, French, and German) of the "Paracetamol" entry from Wikipedia. Additionally, we will change the background of the wordcloud and set a maximum of 50 words.

nltk.download('stopwords')
sw_en = set(nltk.corpus.stopwords.words('english'))
sw_fr = set(nltk.corpus.stopwords.words('french'))
sw_es = set(nltk.corpus.stopwords.words('spanish'))
sw_ca = set(nltk.corpus.stopwords.words('catalan'))
sw_ca.add('efecte')

# At this point, we define the custom stopwords list.

my_stopwords = sw_en.union(sw_fr).union(sw_es).union(sw_ca)

# Retrieve the Wikipedia pages for each language and create a WordCloud.

id_languages = set(["en-simple", "fr", "es", "ca"])
soup = BeautifulSoup(urlopen('http://en.wikipedia.org/wiki/Paracetamol'), 'html.parser')
links = [(el.get('lang'), el.get('title')) for el in soup.select('li.interlanguage-link > a') if el.get('lang') in id_languages]

for language, title in links:
    if language == 'en-simple':
      language='en'
    page_title = title.split(u' – ')[0]
    wikipedia.set_lang(language)
    page = wikipedia.page(page_title)
    print(language)
    print(page.summary)
    print("-----")
    cloud_page_summary = WordCloud(background_color="white", min_word_length=4, max_words=50, stopwords=my_stopwords).generate(page.summary)
    plt.imshow(cloud_page_summary, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
# At this point, we incorporate the libraries that we will need.

from PIL import Image
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
# In order to check the suitability of the image and the color (a black pot on a white background), we look to see if the numbers in the matrix indicate 255.

mask = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/potbo.png'))
mask
```

```python
# There is a possibility that numbers other than 255 may appear. Therefore, we will use "if" and "else" statements along with a "for i in range" loop.

def transform_format(val):
    if np.all (val == 0):
        return 255
    else:
        return val[0]
```

```python
transformed_mask = np.ndarray((mask.shape[0], mask.shape[1]), np.int32)
for i in range(len(mask)):
    transformed_mask[i] = np.array(list(map(transform_format, mask[i])))
```

```python
transformed_mask
```

```python
# This last step is the most complicated one, as it involves the introduction of words into the image.

from nltk.corpus import stopwords

# Download the English stopwords.
nltk.download('stopwords')

# Obtain the English stopwords as a list.
stopwords_en = stopwords.words('english')

article_text1 = BeautifulSoup(urlopen('http://en.wikipedia.org/wiki/Paracetamol'), 'html.parser')
article_text1 = article_text1.get_text()

resumen = hacer_resumen(article_text1)

wc = WordCloud(background_color="white", max_words=50, mask=transformed_mask,
               stopwords=stopwords_en, contour_width=3, contour_color='firebrick')

# Generate a wordcloud
wc.generate(article_text1)

# store to file
wc.to_file('/content/drive/MyDrive/Colab Notebooks/potbo.png')

# show
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
